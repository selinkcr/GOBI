```python
from Bio import SeqIO
import os

def fasta_to_phy(fasta_file, phy_file):
    """
    Converts a FASTA file to a Phylip file using Biopython's SeqIO.

    Parameters:
    fasta_file (str): Path to the input FASTA file.
    phy_file (str): Path to the output Phylip file.
    """

    # Read the FASTA file
    records = list(SeqIO.parse(fasta_file, "fasta"))

    # Write to a Phylip file
    SeqIO.write(records, phy_file, "phylip")

# Example usage
fasta_file = "/content/drive/MyDrive/gobi/deneme/denemeMAFFT.fasta" # Replace with your FASTA file path
phy_file = "/content/drive/MyDrive/gobi/deneme/denemeMAFFT.phy" # Replace with your desired output file path
fasta_to_phy(fasta_file, phy_file)
```

```python
def split_fasta(input_fasta, output_folder):
    """
    Split a FASTA file into multiple FASTA files, each containing one sequence, and save them to a specified folder.

    Parameters:
    - input_fasta: Path to the input FASTA file.
    - output_folder: Path to the folder where output FASTA files will be saved.
    """

    # Ensure the output folder exists, create it if not
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Open the input FASTA file
    with open(input_fasta, 'r') as file:
        content = file.read()

    # Split the file content by '>' to separate sequences, skipping the first empty element if file starts with '>'
    sequences = content.split('>')[1:]

    for seq in sequences:
        # Split each sequence into header and sequence parts
        parts = seq.split('\n', 1)
        header = parts[0].strip().replace(' ', '_').replace('|', '_')
        sequence = parts[1].strip()

        # Create a file name based on the header and ensure it is saved in the output folder
        filename = os.path.join(output_folder, f"{header}.fasta")

        # Write the sequence to its own file
        with open(filename, 'w') as output_file:
            output_file.write(f">{header}\n{sequence}")
        input_fasta = "/content/drive/MyDrive/gobi/Allaligned.fasta" # Replace with your FASTA file path
        output_file = "/content/drive/MyDrive/gobi/selin/output"
        split_fasta(input_fasta_path, output_folder_path)
```

```python
from Bio import AlignIO, SeqIO
from Bio.Align import MultipleSeqAlignment
from Bio.SeqRecord import SeqRecord
from Bio.Seq import Seq


input_file = "/content/drive/MyDrive/gobi/Allaligned.fasta"

alignment = AlignIO.read(input_file, "fasta")

alignment_length = len(alignment[0].seq)
num_seqs = len(alignment)
treshold = 0.3

columns_to_be_deleted = []

# Identify columns for deletion based on first sequence's characters
for column_index in range(alignment_length):
    ref_char = alignment[0].seq[column_index]
    match_count = sum(1 for seq in alignment if seq.seq[column_index] == ref_char)

    # Delete columns if the character from the first sequence is in less than 50% sequences
    if match_count / num_seqs < treshold:
        columns_to_be_deleted.append(column_index)

# Build new sequences excluding the columns marked for deletion
new_records = []
for seq_record in alignment:
    new_seq = ''.join(
        char for idx, char in enumerate(seq_record.seq)
        if idx not in columns_to_be_deleted
    )
    new_records.append(SeqRecord(Seq(new_seq), id=seq_record.id, description=""))

# Save the new alignment to a FASTA file
SeqIO.write(new_records, "filtered_alignment.fasta", "fasta")
```
